```python
def remove_punctuations(text):
    for punctuation in list(string.punctuation):
        text = text.replace(punctuation, '')
    return text
```

```python
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text_data):
    text_data_tokenize=nltk.word_tokenize(text_data)
    text_data_stopwords= [i for i in text_data_tokenize if i not in stopwords]
    text_data=' '.join(text_data_stopwords)
    return text_data
```

```python
''''def clean_number(text):
    text = re.sub(r'(\d+)([a-zA-Z])', '\g<1> \g<2>', text)
    text = re.sub(r'(\d+) (th|st|nd|rd) ', '\g<1>\g<2> ', text)
    text = re.sub(r'(\d+),(\d+)', '\g<1>\g<2>', text)
    return text'''


def remove_digits(text_data):
    text_data_tokenize=nltk.word_tokenize(text_data)
    text_data_tokenize_digits=[data for data in text_data_tokenize if data not in string.digits]
    text_data=' '.join(text_data_tokenize_digits)
    return text_data
```

```python
def clean_text(text):
    text = decontraction(text)
    text=text.lower()
    text = re.sub(r'[^\w\s]','',text, re.UNICODE)
    text = remove_punctuations(text)
    text= remove_stopwords(text)
    text= remove_digits(text)
    return text
```

```python
df['news_short_description'] = df['news_short_description'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

news_categories_values = np.unique(df.news_category.values)

print(news_categories_values)
for category in news_categories_values:
    print('For the news category:', category,'\n')
    print('the top unigrams and bigrams are: \n')
    
    title = '|News Category|Top unigrams|Top Bigrams|'
    subtitle = '|:-----------:|:----------:|:---------:|'
    print(title)
    print(subtitle)


    vect1 = TfidfVectorizer(ngram_range=(1, 1))
    vect2 = TfidfVectorizer(ngram_range=(2, 2)) 
    docs = df.loc[(df.loc[:,'news_category'] ==category),'news_short_description'].values
    top_words = []
    for vect in [vect1, vect2]:
        tfidf = vect.fit_transform(docs)
        sparse_row = vect.transform(['. '.join(docs)])
        scores = dict(zip(vect.vocabulary_, np.array(sparse_row.sum(axis=0))[0]))
        sorted_score = sorted(scores.items(), key=lambda x: x[1], reverse=True)
        top_words.append(', '.join([x[0] for x in sorted_score[:20]]))
    print(f'|{category}|{top_words[0]}|{top_words[1]}|')
    print('\n')
    print('----------------------------------')
```

```python
from sklearn.model_selection import train_test_split
train_input , test_input ,train_output , test_output = train_test_split(df['news_short_description'], df['news_category'] , test_size = 0.2 , random_state = 10)
```

```python
def training_model(train_input,test_input,train_output,test_output,feature_extractor,ml_algorithm):
    
    train_vector= feature_extractor.fit_transform(train_input)
    
    joblib.dump(feature_extractor, str(feature_extractor)+'.pkl')
    
    test_vector = feature_extractor.transform(test_input)
    
    ml_algorithm.fit(train_vector,train_output)
    
    joblib.dump(ml_algorithm, str(ml_algorithm)+str(feature_extractor)+'.pkl')
    
    
    model_prediction=ml_algorithm.predict(test_vector)
    
    return model_prediction,test_output
```

```python
def confusion_matrix(model_prediction,test_output,labels):
    confusion_matrix = metrics.confusion_matrix(test_output, model_prediction)
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix,display_labels=labels)
    cm_display.plot()
    plt.show()
```

```python
def classification_reports(model_prediction,test_output,labels):
    cp_display=classification_report(test_output, model_prediction,target_names=labels)
    print(cp_display)
```

###### Random Forest Classifier, Logistics Regression, Multinomial Naive Bayes using Count Vectorizer and TF-IDF Vectorizer

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

algorithms={'rfc':RandomForestClassifier(),'lr':LogisticRegression(),'mnb':MultinomialNB()}
extractor={'cv':CountVectorizer(),'tfidf_vect':TfidfVectorizer()}

for i in extractor.keys():
    print('FEATURE EXTRACTOR: ',extractor[i])
    print('\n')
    for j in algorithms.keys():
        print('MACHINE LEARNING ALGORITHM: ',algorithms[j])
        model_prediction,test_output=training_model(train_input,test_input,train_output,test_output,extractor[i],algorithms[j])
        print('Accuracy: ',accuracy_score(model_prediction,test_output))
        confusion_matrix(model_prediction,test_output,labels)
        classification_reports(model_prediction,test_output,labels)
        print('\n')
    print('----------------------------------------------------------------\n')
```

###### Stratified K-FOLD

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from statistics import mean, stdev
lr = LogisticRegression()
tfidf = TfidfVectorizer()
list_of_scores=[]
for train_index, test_index in skf.split(df['news_short_description'], df['news_category']):
   x_train, x_test = df.loc[df.index.intersection(train_index),'news_short_description'], df.loc[df.index.intersection(test_index),'news_short_description']
   y_train, y_test = df.loc[df.index.intersection(train_index),'news_category'], df.loc[df.index.intersection(test_index),'news_category']

   
   x_train = tfidf.fit_transform(x_train)
   x_test = tfidf.transform(x_test)
   
   lr.fit(x_train, y_train)
   y_pred = lr.predict(x_test)
   score = accuracy_score(y_test, y_pred)
   list_of_scores.append(score)
print('List of possible accuracy:', list_of_scores)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(list_of_scores)*100, '%')
print('\nMinimum Accuracy:',
      min(list_of_scores)*100, '%')
print('\nOverall Accuracy:',
      mean(list_of_scores)*100, '%')
print('\nStandard Deviation is:', stdev(list_of_scores))
```
